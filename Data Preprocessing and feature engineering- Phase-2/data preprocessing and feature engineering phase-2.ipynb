```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
titanic_df = pd.read_csv("titanic.csv")
```

```python
# Handling Missing Values
# Impute missing values in 'Age' with the median age
titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)
```

```python

# Create a new feature 'HasCabin' and impute missing values in 'Cabin'
titanic_df['HasCabin'] = titanic_df['Cabin'].notnull().astype(int)
```

```python

# Impute missing values in 'Embarked' with the most frequent value
titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0], inplace=True)
```

```python


# Encoding Categorical Variables
titanic_df['Sex'] = titanic_df['Sex'].map({'male': 0, 'female': 1})
titanic_df = pd.get_dummies(titanic_df, columns=['Embarked'], drop_first=True)
```

```python

# Creating New Features
titanic_df['FamilySize'] = titanic_df['SibSp'] + titanic_df['Parch'] + 1
```

```python

# Drop columns that won't be used directly for prediction
titanic_df.drop(['Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
```

```python

# Visualization
# Distribution of Age
plt.figure(figsize=(10, 6))
sns.histplot(titanic_df['Age'], kde=True, bins=30)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python

# Survival rate by Sex
plt.figure(figsize=(10, 6))
sns.barplot(x='Sex', y='Survived', data=titanic_df)
plt.title('Survival Rate by Sex')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.xticks([0, 1], ['Male', 'Female'])
plt.show()
```

```python

# Survival rate by Pclass
plt.figure(figsize=(10, 6))
sns.barplot(x='Pclass', y='Survived', data=titanic_df)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Pclass')
plt.ylabel('Survival Rate')
plt.show()
```

```python

# Survival rate by FamilySize
plt.figure(figsize=(10, 6))
sns.barplot(x='FamilySize', y='Survived', data=titanic_df)
plt.title('Survival Rate by Family Size')
plt.xlabel('FamilySize')
plt.ylabel('Survival Rate')
plt.show()
```

```python


# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(titanic_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
